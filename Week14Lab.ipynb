```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
data_file = r"C:\Users\david\Documents\GitHub\E11_David_Julian_Matt\Week14Data1"
df = pd.read_csv(data_file)
data_file = r"C:\Users\david\Documents\GitHub\E11_David_Julian_Matt\Week14Data2"
df2 = pd.read_csv(data_file)
```

```python
temp = []
for i in df["Air Quality PM 2.5 (ug/m3)"]:
    temp.append(i)
for i in df2["Air Quality PM 2.5 (ug/m3)"]:
    temp.append(i)
plt.hist(temp)
plt.ylabel('Frequency')
plt.xlabel('Counts')
plt.title("Air Quality PM 2.5 (ug/m3)")
plt.show()
print("Mean: " + str(np.mean(temp)))
print("Standard Deviation: " + str(np.std(temp)))
```

Different areas have different levels of pollution, so the air quality tests are not truly random.
